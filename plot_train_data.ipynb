```python
import matplotlib.pyplot as plt
# plot a graph of my training data
# File: 
train_loss_file = "results\Full train(balanced)\model_20230329-155809_train_loss.txt"
train_masked_loss_file = "results\Full train(balanced)\model_20230329-155809_train_masked_loss.txt"
val_loss_file = "results\Full train(balanced)\model_20230329-155809_val_loss.txt"
val_masked_loss_file = "results\Full train(balanced)\model_20230329-155809_val_masked_loss.txt"
```

```python
# plot train loss
training_losses = []
with open(train_loss_file, 'r') as f:
    for line in f:
        training_losses.append(float(line))
epochs = range(1, len(training_losses) + 1)
fig = plt.figure(figsize=(7,6))
plt.plot(epochs, training_losses, 'b', label='Training Loss')
plt.title('Training UNet: Train Set Losses')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
plt.tight_layout()
# save plot as pdf
fig.savefig('figs\\training\\train_loss.pdf', format='pdf')
```

```python
# plot trainiing masked loss
training_masked_losses = []
with open(train_masked_loss_file, 'r') as f:
    for line in f:
        training_masked_losses.append(float(line))
epochs = range(1, len(training_masked_losses) + 1)
fig = plt.figure(figsize=(7,6))
plt.plot(epochs, training_masked_losses, 'b', label='Training UNet Masked Losses')
plt.title('Training UNet: Train Set Masked Losses')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
plt.tight_layout()
# save plot as pdf
fig.savefig('figs\\training\\train_masked_loss.pdf', format='pdf')
```

```python
# plot validation loss
validation_losses = []
with open(val_loss_file, 'r') as f:
    for line in f:
        validation_losses.append(float(line))
epochs = range(1, len(validation_losses) + 1)
fig = plt.figure(figsize=(7,6))
plt.plot(epochs, validation_losses, 'b', label='Validation Losses')
plt.title('Training UNet: Validation Set Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
plt.tight_layout()
# save plot as pdf
fig.savefig('figs\\training\\val_loss.pdf', format='pdf')
```

```python
# plot validation masked loss
validation_masked_losses = []
with open(val_masked_loss_file, 'r') as f:
    for line in f:
        validation_masked_losses.append(float(line))
epochs = range(1, len(validation_masked_losses) + 1)
fig = plt.figure(figsize=(7,6))
plt.plot(epochs, validation_masked_losses, 'b', label='Validation Masked Loss')
plt.title('Training UNet: Validation Set Masked Losses')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid()
# make y axis have a similar scale to the plot above
plt.ylim(0, 0.025)

plt.legend()
plt.show()

plt.tight_layout()
# save plot as pdf
fig.savefig('figs\\training\\val_masked_loss.pdf', format='pdf')
```
