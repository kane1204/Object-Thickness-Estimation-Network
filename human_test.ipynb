```python
from os import listdir
from os.path import isfile, join
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplcursors
from PIL import Image
dir = "data\human_samples"
```

```python
# Read in all the images and csv file and randomly label the images with a red or blue dot
import ast
 
imgs = []
points =  [[],[]]*10
# read in csv file
real_df = pd.read_csv("human_sample_actuals.csv", index_col=0)
```

```python
label_thicker = {}
real_df['label'] = None
real_df['image_name'] = None
for num, data in real_df.iterrows():
    img_file = join(dir, data["file_name"], "img.png")
    # pick which point will be red and which will be blue
    if np.random.randint(2) == 0:
        redpt = data["thicker"]
        bluept = data["thinner"]
        label_thicker[num] = "red"
    else:
        redpt = data["thinner"]
        bluept = data["thicker"]
        label_thicker[num] = "blue"
    
    #  append the label to the dataframe
    real_df["label"].iloc[num] = label_thicker[num]
    real_df["image_name"].iloc[num] = str(num)+"_labeled.png"
    # real_df['label'] = label_thicker[num]
    # read in the image and add the red and blue dots
    redpt = ast.literal_eval(redpt)
    bluept =  ast.literal_eval(bluept)

    image = Image.open(img_file)
    # add red dot
    image.putpixel((redpt[1],redpt[0]), (255,0,0))
    image.putpixel((redpt[1]+1,redpt[0]), (255,0,0))
    image.putpixel((redpt[1]-1,redpt[0]), (255,0,0))
    image.putpixel((redpt[1],redpt[0]+1), (255,0,0))
    image.putpixel((redpt[1],redpt[0]-1), (255,0,0))
    # add a larger blue dot
    image.putpixel((bluept[1],bluept[0]), (0,0,255))
    image.putpixel((bluept[1]+1,bluept[0]), (0,0,255))
    image.putpixel((bluept[1]-1,bluept[0]), (0,0,255))
    image.putpixel((bluept[1],bluept[0]+1), (0,0,255))
    image.putpixel((bluept[1],bluept[0]-1), (0,0,255))
    # save the image
    image.save(join(dir,"labeled_samples", str(num)+"_labeled.png"))
    image.close()
print(label_thicker)
real_df.to_csv("human_sample_actuals.csv")
```
